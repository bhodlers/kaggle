```python
import torch
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

model = LogisticRegression()
model.fit(X, y)

import joblib
joblib.dump(model, "/kaggle/working/model.pkl")
```
